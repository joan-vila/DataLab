```python
import numpy as np

array_1d = np.array([1, 2, 3])
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
print(array_1d[0])
print(array_2d[0, 0])
print(array_2d[0:1, :])
print(array_2d[:, 1])

print(np.mean(array_1d))
print(np.median(array_1d))
print(np.std(array_1d))
```

```python
# customizing the plot
import matplotlib.pyplot as plt
import numpy as np
year = [x for x in range(1950, 2101)]
pop = np.random.randint(1, 10, 151)
plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks([0, 2, 4, 6, 8, 10], ['0', '2B', '4B', '6B', '8B', '10B'])
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
# x arrey of length 1000 with random integers between 1 and 6
x = np.random.randint(1, 7, 1000)
y = []
# y must be equal to the arithmetic mean of x
sum = 0
for i in range(0, len(x)):
    sum += x[i]
    y.append(sum/(i+1))

plt.plot(y)    
plt.show()
```
